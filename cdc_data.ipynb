```python
#coded by Stacey D.
#import dependencies
import pandas as pd
```

```python
# Reference the file where the CSV is located
file = "Resources/CDC-covid.csv"

# Import the data into a Pandas DataFrame
cdc_data = pd.read_csv(file)
cdc_data
```

```python
# Get a list of all of our columns for easy reference
cdc_data.columns
```

```python
# Extract "Date As Of", "Start Date", "End Date", "Group", "Year", "Month",
# "State", "Sex", "Age Group", "COVID-19 Deaths", and "Total Deaths"
reduced_cdc_data = cdc_data.loc[:, ['Data As Of', 'Start Date', 'End Date', 'Group', 'Year', 'Month',
       'State', 'Sex', 'Age Group', 'COVID-19 Deaths', 'Total Deaths']]
reduced_cdc_data
```

```python
#drop rows with NaN
reduced_cdc_data = reduced_cdc_data.dropna()
```

```python
#we ony want ages 18 and up, because people under the age of 18 can't spend money on the stock market
unique_ages = reduced_cdc_data['Age Group'].unique()
unique_ages

#move age group to index row
reduced_cdc_data = reduced_cdc_data.set_index("Age Group")
reduced_cdc_data

#use only 18+
reduced_cdc_data = reduced_cdc_data.loc[['18-29 years',
       '30-39 years','40-49 years',
       '50-64 years', '65-74 years', '75-84 years',
       '85 years and over'], : ]
```

```python
#removed rows where covid deaths = 0
cdc_data_df = reduced_cdc_data[ reduced_cdc_data['COVID-19 Deaths'] != 0 ]
cdc_data_df
```

```python
#export to output folder
cdc_data_df.to_csv("Output/cdc_data_df.csv", index=False, header=True)
```
